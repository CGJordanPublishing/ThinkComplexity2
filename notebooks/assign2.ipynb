```python
# Cameron Jordan, 100377297
import gzip
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
from utils import decorate
try:
    import empiricaldist
except ImportError:
    !pip install empiricaldist
from empiricaldist import Pmf, Cdf

def all_pairs(nodes):
    for i, u in enumerate(nodes):
        for j, v in enumerate(nodes):
            if i < j:
                yield u, v

def read_actor_network(filename, n=None):
    G=nx.Graph()
    with gzip.open(filename) as f:
        for i, line in enumerate(f):
            nodes=[int(x) for x in line.split()]
            G.add_edges_from(all_pairs(nodes),)
            if n and i >= n:
                break
    return G
G=read_actor_network("actor.dat.gz")
```

```python
#average degree
def degrees(G):
    return [G.degree(u) for u in G]
degmean=np.mean(degrees(G))
degmean
```

```python
#number of actors
len(degrees(G))
```

```python
pmf=Pmf.from_seq(degrees(G))
cdf=Cdf.from_seq(degrees(G))
```

```python
#pmf plot
plt.figure(figsize=(8,4))
options = dict(ls='', marker='.')
pmf.plot(label='actor', color='C0', **options)
decorate(xscale='log', yscale='log',
         xlabel='Degree', ylabel='PMF')
```

```python
#cdfplot log-x scale
plt.figure(figsize=(8,4))
cdf.plot(color='C0')
decorate(xscale='log', xlabel='Degree', ylabel='CDF')
```

```python
#cdfplot log-log scale
plt.figure(figsize=(8,4))
cdf.plot(color='C0')
decorate(xscale='log', yscale='log', xlabel='Degree', ylabel='CDF')
```
